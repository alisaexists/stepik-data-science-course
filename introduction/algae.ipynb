```python
import pandas as pd
```

```python
df = pd.read_csv('algae.csv')
```

```python
df.head()
```

```python
df[df['species'].str.contains("Fucus")]
```

```python
df[df['genus'] == 'Fucus'].describe()
```

```python
df.loc[df.genus == 'Fucus', 'alanin'].agg(['min', 'mean', 'max']).round(2)
```

```python
df.groupby('group').agg({'citrate': 'var', 'sucrose': ['min', 'max'], })
```

```python
df.groupby('group') \
    .agg({'citrate': 'var', 'sucrose': lambda x: x.max() - x.min(), 'glucose': 'count'}) \
    .round(2)
```

```python
import numpy as np
df.groupby('group').aggregate({"species":'nunique', 'citrate':'var','sucrose': np.ptp})
```
