```python
import pandas as pd
import numpy as np
```

```python
invasion_data = pd.read_csv('invasion.csv')
```

```python
invasion_data.head()
```

```python
invasion_data.shape
```

```python
X = invasion_data
X = X.drop('class', axis=1)
X.head()
```

```python
y = invasion_data['class']
```

```python
y.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf = RandomForestClassifier(random_state=0)
```

```python
params = {'n_estimators': range(10, 51, 10),
          'max_depth': range(1, 12, 2),
          'min_samples_leaf': range(1, 8),
          'min_samples_split': range(2, 10, 2)}
```

```python
gridsearc_clf = GridSearchCV(clf, params, cv=3, n_jobs=-1)
```

```python
gridsearc_clf.fit(X, y)
```

```python
best_clf = gridsearc_clf.best_estimator_
```

```python
X_test = pd.read_csv('operative_information.csv')
```

```python
y_pred = best_clf.predict(X_test)
```

```python
y_pred
```

```python
unique, counts = np.unique(y_pred, return_counts=True)
```

```python
feature_importances = best_clf.feature_importances_
feature_importances_df = pd.DataFrame({'features': X.columns,
                                       'feature_importances': feature_importances})
```
